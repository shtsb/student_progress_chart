```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

students_and_grades = {
    "math" : [19,18,13,16,11] ,
    "physics":[10,12,20,12,19.5] ,
    "history":[5,16,9,14,18.5] ,
    "chemistry" :[20,14,13,16,11] ,
    "geography" : [12,12,8,13,16]
}

print("The grades of the students in this class are equal to :\n")
a = pd.DataFrame(students_and_grades,index = ["lilli","ella","emma","mia","nina"])
print(a,"\n date : 2025/4/28\n\n")



students_and_grades = {
    "math" : [17,19,15,20,15] ,
    "physics":[13,14,19,17,20] ,
    "history":[11,19,14,13,17.5] ,
    "chemistry" :[20,13,14,18,13] ,
    "geography" : [15,14,11,15,16]
}

print("The grades of the students in this class are equal to :\n")
b = pd.DataFrame(students_and_grades,index = ["lilli","ella","emma","mia","nina"])
print(b,"\n date : 2025/6/28")

plt.figure(figsize=(17,8))
lesson = np.array (['math','physics','history','chemistry','geography'])

emma_grades1 = np.array([13,20,9,13,8])
emma_grades2 = np.array([15,19,14,14,11])
plt.subplot(2,3,1)
plt.plot(lesson,emma_grades1,c='#808080')
plt.plot(lesson,emma_grades2 , c= '#008000')
plt.title("emma")



lilli_grades1 = np.array([19,10,5,20,12])
lilli_grades2 = np.array([17,13,11,20,15])
plt.subplot(2,3,2)
plt.plot(lesson,lilli_grades1,c='#808080')
plt.plot(lesson,lilli_grades2 , c='#008000')
plt.title("lilli")




ella_grades1 = np.array([18,12,16,14,12])
ella_grades2 = np.array([19,14,19,13,14])
plt.subplot(2,3,3)
plt.plot(lesson,ella_grades1,c='#808080')
plt.plot(lesson,ella_grades2 , c='#008000')
plt.title("ella")




mia_grades1 = np.array([19,16,14,16,13])
mia_grades2 = np.array([20,17,13,18,15])
plt.subplot(2,3,4)
plt.plot(lesson,ella_grades1,c='#808080')
plt.plot(lesson,ella_grades2 , c='#008000')
plt.title("mia")



nina_grades1 = np.array([11,19.5,18.5,11,16])
nina_grades2 = np.array([15,20,17.5,13,16])
plt.subplot(2,3,5)
plt.plot(lesson,nina_grades1,c='#808080')
plt.plot(lesson,nina_grades2 , c='#008000')
plt.title("nina")

plt.suptitle("Student progress chart")
plt.show()
```
